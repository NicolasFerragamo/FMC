# <span style="color:blue"> Propiedades de la potenciación <span>
  
##  <span style="color:blue">Introducción </span>    
En esta sección vamos a repasar los conceptos matemáticos que nos van a permitir usar la notación científica.

<div class="alert alert-block alert-warning">
 <b> La potenciación es una operación matemática entre dos términos denominados: base ${\displaystyle a}$ y exponente ${\displaystyle n}$. Se escribe $a^n$ y se lee normalmente como "a elevado a la n"</b> 
</div>


<div class="alert alert-block alert-warning">
 <b> Importante si bien se puede trabajar con exponentes racionales e irracionales, nosotros solo vamos a trabajar con exponentes enteros </b>
</div>
    
##  <span style="color:blue">Propiedades </span>
    
<div class="alert alert-block alert-info">
 <b> Exponente entero</b> 
</div>  
    
Elevar un número <b>$a$</b> a un exponente <b>$ n \in \mathbb N$</b> es multiplicar <b>$a$</b> por si mismo <b>$n$</b> veces.

<div class="alert alert-block alert-info">

$ a^1 = a $
    
$ a^2 = a\hspace{10px} x \hspace{10px}a $  
    
$ a^3 = a\hspace{10px} x \hspace{10px}a \hspace{10px} x \hspace{10px}a $ 
    
$ \vdots   \hspace{50px}  \vdots $   
    
$ a^n = \underbrace{a\hspace{10px}x \hspace{10px}a\hspace{10px} x\hspace{10px} a\hspace{10px} x... a}_{n 
\hspace{10px}veces} $

</div>  
    
<div class="alert alert-block alert-success">
 <b> Por ejemplo</b> 

$ 2^1 = 2 $
    
$ 2^2 = 2\hspace{10px} x \hspace{10px}2 = 4 $
    
$ 2^3 = 2\hspace{10px} x \hspace{10px}2 \hspace{10px} 2 \hspace{10px}2 = 8 $$    
</div>
    
Observe algo muy importante cuando paso de $2^3$ a $2^2$ lo que estamos haciendo es dividir por 2, $8 \div 2 = 4$. Lo mismo ocurre cuando pasamos de $2^2$ a $2^1$, seguimos dividiendo por 2, $4 \div 2 = 2$. Ahora bien nada nos impide suponer que lo mismo ocurre cuando pasamos de $2^1$ a $2^0$ por lo tanto $2 \div 2 = 1$. Por lo cual **¡acabamos de demostrar por que $a^0 = 1$ simpre !!** Pero la cosa no termina aca, podemos seguir un poco mas, de la misma forma si hacemos $1 \div 2 = \frac{1}{2}$ por lo que pasamos de $ 2^0$ a $2^{-1} = \frac{1}{2}$. 

<div class="alert alert-block alert-info">   
<b>Por definición $$a^0 = 1$$</b>    
</div>

<div class="alert alert-block alert-info">   
<b>Producto de potencias de igual base</b>    
</div>

¿Qué pasa cuando multiplicamos dos o mas números con la misma base a distinto exponente? En este caso se utiliza la misma base y se suman los exponentes.

<div class="alert alert-block alert-info">   
<b>$$a^m\hspace{10px} x \hspace{10px}a^n = a^{m+n} $$</b>    
</div>

Si lo piensan es muy lógico, estamos multiplicando 2 números y cada número se multiplica a si mismo tantas veces como indique el exponete, por lo cual al descomponer la potencia quedan los **m** numeros multiplicando a los **n** números, por lo cual tenenmos **m+n** números multilicandose.

<div class="alert alert-block alert-success">
<b> Como siempre veamos un ejemplo:</b> 

$$ 3^5 \hspace{10px} x \hspace{10px} 3^3 = 3^8 $$
    
si descomponemos ambas potencias nos queda
    
$$ (3*3*3*3*3) \hspace{10px} x \hspace{10px} (3*3*3)$$
    
En total tenemos ocho 3 multiplicándose a si mismos, por lo cual si los juntamos en un solo exponente nos queda
    
$$ 3^8 $$    
</div>


<div class="alert alert-block alert-info">   
<b>Potencia de una potencia</b>    
</div>

¿Qué pasa cuando elevamos un número **a** a una potencia **n** y luego a otra potencia **m**? En este caso se multiplican los exponentes.

<div class="alert alert-block alert-info">   
<b>$$ (a^m)^n = a^{m.n} $$</b>    
</div>

Si lo piensan esto también es muy lógico, bueno la matemática se basa en la lógica. Estamos multiplicando la base **m** veses y ese resultado se multiplica **n** veces.

<div class="alert alert-block alert-success">
<b> Por ejemplo:</b> 

$$ (3^2)^4 = 3^8 $$
    
si descomponemos por parte nos queda que tenemos que multiplicar $3^2$ cuatro veces:
    
$$ (3^2) \hspace{10px} x \hspace{10px} (3^2) \hspace{10px} x \hspace{10px} (3^2)\hspace{10px} x \hspace{10px} (3^2)$$
    
$$ (3*3) \hspace{10px} x \hspace{10px} (3*3) \hspace{10px} x \hspace{10px} (3*3)\hspace{10px} x \hspace{10px} (3*3) $$    
    
En total tenemos ocho 3 multiplicándose a si mismos, por lo cual si los juntamos en un solo exponente nos queda
    
$$ 3^8 $$    
</div>

<div class="alert alert-block alert-info">   
<b>Potencia de un producto</b>    
</div>

¿Qué pasa cuando multiplicamos un número **a** por un número **b** elevado a la **n**? En este caso la potencia de un Producto es igual al producto de cada uno de los factores elevado al mismo exponente, es decir: 

<div class="alert alert-block alert-info">   
<b>$$(a * b)^n = a^n * b^n $$</b>    
</div>

Esta propiedad se deriva de que es lo mismo multiplicar $(a*b)$ primero y al resultado multiplicarlo **n** veces que multiplicar **n** veces a y **n** veces b y luego multiplicar ambos resultados.

<div class="alert alert-block alert-success">
<b> Por ejemplo:</b> 

$$ (4*5)^2 = 400 = (4^2 * 5^2) = 16 * 25 = 400 $$
    
si resolvemos primero la potencia nos queda:
    
$$ (4*5) \hspace{10px} x \hspace{10px} (4*5)$$
    
Si utilizamos la propiedad asociativa lo podemos ordenar de la siguietne forma:    
    
$$ (4*4) \hspace{10px} x \hspace{10px} (5*5) $$    
    
Por último podemos escribirlo en forma de potencia quedando
    
$$ 4^2*5^2 $$    
</div>

<div class="alert alert-block alert-info">   
<b>Potencia de un número negativo</b>    
</div>

¿Qué pasa cuando la base es negativa? La verdad no mucho solo hay que acordarse la regla de los signo: 

<div class="alert alert-block alert-warning">   
<b>$$+ * + = +$$ 
    
$$+ * - = -$$
    
$$- * + = -$$
    
$$- * - = +$$</b> 
</div>

Entonces si el exponente es par el resultado es positivo y si el exponente es impar el resultado es negativo.

<div class="alert alert-block alert-danger">   
Los números pares se representan como <b>2n</b> y los impares como <b>2n + 1</b> 
</div>
<div class="alert alert-block alert-info">   
<b>$$(-a^n) = a^n si \hspace{10px} n \hspace{10px}es\hspace{10px} par$$</b>  
    
<b>$$(-a^n) = -a^n si \hspace{10px} n \hspace{10px}es \hspace{10px}par$$</b>      
</div>

Esta propiedad se deriva de que si multiplicamos (-1) un número par de veces el resultado es (+1) (ver regla de los signos) y si se multiplica un número impar de veces nos termina quedando un $ + * - = -$

<div class="alert alert-block alert-success">
<b> Ejemplo de potencia par:</b> 

$$ (-8)^4 = 8^4 $$
    
si resolvemos la potencia nos queda:
    
$$ (-8) \hspace{10px} x \hspace{10px} (-8) \hspace{10px} x \hspace{10px} (-8) \hspace{10px} x \hspace{10px} (-8)$$
    
Si utilizamos la propiedad asociativa lo podemos ordenar de la siguietne forma:    
    
$$ (-8*-8) \hspace{10px} x \hspace{10px} (-8*-8) $$    
    
$$ (64) \hspace{10px} x \hspace{10px} (64) $$ 
    
Por último podemos escribirlo en forma de potencia quedando
    
$$ 4096 $$    
</div>

<div class="alert alert-block alert-success">
<b> Ejemplo de potencia impar:</b> 

$$ (-8)^5 = -8^5 $$
    
si resolvemos la potencia nos queda:
    
$$ (-8) \hspace{10px} x \hspace{10px} (-8) \hspace{10px} x \hspace{10px} (-8) \hspace{10px} x \hspace{10px} (-8) \hspace{10px} x \hspace{10px} (-8)$$
    
Si utilizamos la propiedad asociativa lo podemos ordenar de la siguietne forma:    
    
$$ (-8*-8) \hspace{10px} x \hspace{10px} (-8*-8) \hspace{10px} x \hspace{10px} (-8) $$    
    
$$ (64) \hspace{10px} x \hspace{10px} (64) \hspace{10px} x \hspace{10px} (-8)$$ 
    
Por último podemos escribirlo en forma de potencia quedando
    
$$ -32768$$    
</div>

<div class="alert alert-block alert-info">   
<b>Inverso multiplicativo</b>    
</div>

¿Qué pasa cuando el exponente es negativa?  El exponente negativo solo indica que hay que cambiar la base por su inverso multiplicativo (si esta multiplicando dividrilo por ese número.)

Esta propiedad se deriva de lo que vimos con el ejemplo de base 2.

<div class="alert alert-block alert-info">   
<b>$$a^{-1} = \frac{1}{a}$$</b>   
    
Si el exponente es distinta de 1  
    
    
<b>$$a^{-n} = \frac{1}{a^n}$$</b>      
</div>


<div class="alert alert-block alert-success">
<b> Ejemplo de exponente unitario:</b> 
    
<b>$$5^{-1} = \frac{1}{5}=0,2$$</b>    
</div>

<div class="alert alert-block alert-danger">
<b> Note algo fundamental, si el exponente es negativo el número que representa es menor a 1 $\Rightarrow$ podemos decir que un exponente negativo representa números pequeños y un exponente positivo representa números grandes</b> 
</div>

<div class="alert alert-block alert-success">
<b> Ejemplo de exponente no unitario:</b> 

$$ (2)^{-3}= \frac{1}{2^3} $$
    
si resolvemos la potencia nos queda:
    
$$ (2)^{-3}= \frac{1}{2} \hspace{10px} x \hspace{10px} \frac{1}{2} \hspace{10px} x \hspace{10px}\frac{1}{2} $$
    
$$ \frac{1}{8} = 0,125$$  
</div>

No olvide que la base también puede ser negativa y no hay ningún cambio.


<div class="alert alert-block alert-success">
<b> Ejemplo de exponente no unitario y base negativa:</b> 
    
como el exponente es impar la base queda negativa.
    
$$ (-4)^{-3}= -\frac{1}{4^3} $$
    
si resolvemos la potencia nos queda:
    
$$ \frac{1}{64} = -0,015625$$ 
</div>

<div class="alert alert-block alert-info">   
<b>División de potencias de igual base</b>    
</div>

¿Qué pasa cuando hay una fracción?  Para resolver esto tenemos que usar lo que ya aprendimos, sabemos que el denominador de la fracción lo puedo escribir como un exponente negativo en el numerador. Vamos a usar eso y luego nos quedan dos números de igual base multiplicándose, por lo cual como ya vimos hay que sumar los exponentes. **Pero ojo porque un exponente es negativo por lo cual en realidad se restan**.


<div class="alert alert-block alert-info">   
<b>$$\frac{a^m}{a^n} = a^m * a^{-n} = a^{m-n}$$</b>     
    
$$\boxed{\frac{a^m}{a^n} = a^{m-n}}$$      
</div>


<div class="alert alert-block alert-success">
<b> Ejemplo:</b> 

<b>$$\frac{2^5}{2^3}$$</b>  
    
Empesemos utilizando la propiedad del **Inverso multiplicativo** al segundo término
    
<b>$$ 2^5 * 2^{-3}$$</b>  
    
Aplicando la propiedad de Producto de potencias de igual base
    
<b >$$ 2^{5-3} = 2^2 = 4 $$</b>   
    
De otra forma si resolvemos primero las potencias lo podemos ver como:
    
$$\frac{2^5}{2^3} = \frac{\overbrace{2*2*2*2*2}^{5}}{\underbrace{2*2*2}_{3}} = 2*2 = 4$$
</div>





<div class="alert alert-block alert-info">   
<b>Potencia de una fracción</b>    
</div>

¿Qué pasa cuando tenemos una fracción elevada a una potencia?  A esta altura del apunte vamos a ser breves La potencia de un cociente es igual al cociente de cada uno de los números elevado al mismo exponente. 

<div class="alert alert-block alert-info">   
<b>$$ \left( \frac{a}{b} \right)^n = \frac{a^n}{b^n}$$</b>       
</div>


<div class="alert alert-block alert-success">
<b> Ejemplo de exponente unitario:</b> 
    
<b>$$ \left( \frac{8}{4} \right)^2 $$</b>    
    
Podemos resolver la potencia de la siguiente manera:
    
<b>$$ \left( \frac{8}{4} \right) \hspace{10px} x \hspace{10px} \left( \frac{8}{4} \right) $$</b>  
    
Podemos reordenar de la siguinete manera
    
<b>$$ \frac{8*8}{4*4}  $$</b>  

y obtenemos lo que queríamos
    
<b>$$ \frac{8^2}{4^2}  $$</b>     
</div>
